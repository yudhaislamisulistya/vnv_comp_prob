```python
import numpy as np ## Import numpy untuk menghitung matriks
import random ## Import random untuk memilih path secara random
import math ## Import math untuk menghitung probabilitas Poisson
import pandas as pd ## Import pandas untuk menampilkan matriks
from IPython.display import display ## Import display untuk menampilkan matriks

U = [] ## Matriks probabilitas
def generate_stochastic_matrix(n: int, m: int) -> np.ndarray: ## Fungsi untuk menghitung matriks probabilitas
    matrix = np.random.rand(n, m) ## Matriks probabilitas
    matrix /= matrix.sum(axis=1)[:, np.newaxis] ## Matriks probabilitas
    print("Martix Probabilitas Stochastic:") ## Menampilkan matriks probabilitas
    display(pd.DataFrame(matrix)) ## Menampilkan matriks probabilitas
    return matrix ## Mengembalikan nilai matriks probabilitas


def P(σ: list[int]) -> float: ## Fungsi untuk menghitung probabilitas path
    # U = np.array([[0.7, 0.3, 0.0],[0.0, 0.4, 0.6],[0.0, 0.0, 1.0]]) ## Matriks probabilitas yang sudah ditentukan
    P_sigma = 1 ## Variabel untuk menampung probabilitas path
    for i in range(len(σ) - 1): ## Perulangan untuk menghitung probabilitas path
        P_sigma *= U[σ[i], σ[i+1]] ## Rumus probabilitas path
    return P_sigma ## Mengembalikan nilai probabilitas path

def PP(lambda_t: float, n: int) -> float: ## Fungsi untuk menghitung probabilitas Poisson
    return math.exp(-lambda_t) * (lambda_t ** n) / math.factorial(n) ## Rumus probabilitas Poisson

def Pr(Yt: float, y: float, k_sigma: float) -> float: # Fungsi untuk menghitung probabilitas reward
    if Yt <= y: # Jika nilai Yt kurang dari atau sama dengan y
        return k_sigma  # Mengembalikan nilai k_sigma
    else: # Jika nilai Yt lebih dari y
        return 1 - k_sigma # Mengembalikan nilai 1 - k_sigma
    
def last(σ): ## Fungsi untuk mengambil nilai terakhir dari path
    return σ[-1] ## Mengembalikan nilai terakhir dari path

def first(σ): ## Fungsi untuk mengambil nilai pertama dari path
    return σ[0] ## Mengembalikan nilai pertama dari path

def Y(t, y): ## Fungsi untuk menghitung probabilitas dan error
    s = np.arange(y).tolist() ## State awal
    s_aksen = y - 1 ## State akhir
    prob = 0 # Variabel untuk menampung probabilitas
    error = 0 # Variabel untuk menampung error
    paths = [s] # Variabel untuk menampung path
    print("State             : " , s)
    print("State Akhir       : " , s_aksen)
    
    # Menentukan Variabel w dan N untuk menghitung probabilitas
    w = 0.02 # Variabel untuk menampung nilai w
    N = 50 # Variabel untuk menampung nilai N
    lambda_t = 0.5 * t # Variabel untuk menampung nilai lambda_t
    k_sigma = 0.5 # Variabel untuk menampung nilai k_sigma
    print("w                 : " , w)
    print("N                 : " , N)
    print("lambda_t (0.5*t)  : " , lambda_t)
    print("k_sigma           : " , k_sigma)
    print("t                 : " , t)
    print("y                 : " , y)
    
    i = 0 # Variabel untuk menampung perulangan
    while paths != []: ## Perulangan untuk menghitung probabilitas dan error
        σ = random.choice(paths) # Memilih path secara random
        paths.remove(σ) # Menghapus path yang sudah dipilih
        if P(σ) > w and np.sum(σ) < N: ## Jika probabilitas path lebih besar dari w dan jumlah path kurang dari N
            if(last(σ) == s_aksen): ## Jika nilai terakhir path sama dengan state akhir
                prob +=  PP(lambda_t, np.sum(σ)) * P(σ) * Pr(t, y, k_sigma) ## Rumus probabilitas
                print('\n')
                print("Perulangan ke          : " , (i+1))
                print("Path                   : " , σ)
                print("Probabilitas Path      : " , P(σ))
                print("Probabilitas Poisson   : " , PP(lambda_t, np.sum(σ)))
                print("Probabilitas Reward    : " , Pr(t, y, k_sigma))
                print("Probabilitas Total     : " , prob)
                i += 1 ## Menambahkan nilai perulangan
            for s_prime in s: ## Perulangan untuk menambahkan path baru
                new_path = σ + [s_prime,] ## Menambahkan path baru
                paths.append(new_path) ## Menambahkan path baru ke dalam variabel paths
        else: ## Jika probabilitas path kurang dari w atau jumlah path lebih dari N
            for n in range(len(σ) - 1): ## Perulangan untuk menghitung error
                error += (1 - PP(lambda_t, n)) * P(σ) ## Rumus error
                # print("Error Path ke-" , (n+1) , " : " , error)
                
                
    return prob, error ## Mengembalikan nilai probabilitas dan error


t = input("Masukkan nilai t: ") ## Input nilai t
y = input("Masukkan nilai y: ") ## Input nilai y
t = int(t) ## Mengubah nilai t menjadi integer
y = int(y) ## Mengubah nilai y menjadi integer

U = generate_stochastic_matrix(y, y) ## Memanggil fungsi generate_stochastic_matrix untuk menghitung matriks probabilitas
prob, error = Y(t, y) ## Memanggil fungsi Y untuk menghitung probabilitas dan error
print('\n')
print("Nilai Probabilitas SUM : ", prob) ## Menampilkan nilai probabilitas
print("Nilai Error        SUM : ", error) ## Menampilkan nilai error
```
